```python
import pandas as pd
import numpy as np
```

```python
my_df = pd.read_csv('fci-breeds.csv')
my_df
```

```python
#counting distinct value in DESCENDING ORDER
my_df['group'].value_counts()
```

```python
#counting distinct value in ASCENDING ORDER
my_df['group'].value_counts(ascending=True)
```

```python
#dog names without NAN(Not A Number)
my_df['name'].value_counts()
```

```python
#dog names with NAN
my_df['name'].value_counts(dropna=False)
```

```python
#get relative frequency - percentage
my_df['name'].value_counts(normalize=True)
```

```python
#get specific item count
#my_df['name'].value_counts()['PUG']
my_df['group'].value_counts()['Retrievers']
```

```python
#count unique values - size
my_df.groupby('name').size()
```

```python
#count uinque values - count
my_df.groupby('name').count()
```

```python
#get a count of all columns across all columns
my_df.apply(pd.value_counts)
```
